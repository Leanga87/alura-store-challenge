```python
import csv
import requests
from io import StringIO

# Función para leer los datos de un CSV desde una URL
def leer_csv(url):
    response = requests.get(url)
    data = response.text
    f = StringIO(data)
    reader = csv.DictReader(f)
    return [row for row in reader]  # Devuelve una lista de diccionarios

# URLs de los archivos CSV
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Cargar los datos correctamente
tienda_1 = leer_csv(url1)
tienda_2 = leer_csv(url2)
tienda_3 = leer_csv(url3)
tienda_4 = leer_csv(url4)

# Verificar las primeras filas de los datos
print("Encabezados de la tienda 1:", tienda_1[0].keys())  # Mostrar claves de la primera fila
print("Encabezados de la tienda 2:", tienda_2[0].keys())
print("Encabezados de la tienda 3:", tienda_3[0].keys())
print("Encabezados de la tienda 4:", tienda_4[0].keys())
```

# 3. Calificación promedio de la tienda


```python
# Función para calcular la calificación promedio por tienda
def calcular_calificacion_promedio_cliente(tienda):
    total_calificaciones = 0
    total_resenas = 0

    # Iterar sobre los datos de la tienda
    for fila in tienda:
        calificacion = float(fila['Calificación'])  # Obtener la calificación como número flotante
        total_calificaciones += calificacion  # Sumar la calificación
        total_resenas += 1  # Contar el número de reseñas

    if total_resenas > 0:
        return total_calificaciones / total_resenas  # Calcular promedio
    else:
        return 0  # Si no hay reseñas, el promedio es 0

# Calcular la calificación promedio de cada tienda
calificacion_promedio_tienda_1 = calcular_calificacion_promedio_cliente(tienda_1)
calificacion_promedio_tienda_2 = calcular_calificacion_promedio_cliente(tienda_2)
calificacion_promedio_tienda_3 = calcular_calificacion_promedio_cliente(tienda_3)
calificacion_promedio_tienda_4 = calcular_calificacion_promedio_cliente(tienda_4)
```

```python
print(f"Calificación promedio tienda 1: {calificacion_promedio_tienda_1:.2f}")
print(f"Calificación promedio tienda 2: {calificacion_promedio_tienda_2:.2f}")
print(f"Calificación promedio tienda 3: {calificacion_promedio_tienda_3:.2f}")
print(f"Calificación promedio tienda 4: {calificacion_promedio_tienda_4:.2f}")
```

# 4. Productos más y menos vendidos

```python
# Función para identificar los productos más y menos vendidos
def productos_mas_y_menos_vendidos(tienda):
    # Crear un diccionario para contar las ventas de cada producto
    ventas = {}

    # Recorrer todos los productos en la tienda
    for fila in tienda:
        producto = fila['Producto']  # Obtener el nombre del producto
        # Si el producto ya está en el diccionario, sumamos 1 a su contador
        if producto in ventas:
            ventas[producto] += 1
        else:
            ventas[producto] = 1  # Si el producto no está, lo agregamos con un contador de 1

    # Encontrar el producto más vendido (el de mayor valor en el diccionario)
    producto_mas_vendido = max(ventas, key=ventas.get)  # Clave con el valor máximo
    ventas_producto_mas_vendido = ventas[producto_mas_vendido]

    # Encontrar el producto menos vendido (el de menor valor en el diccionario)
    producto_menos_vendido = min(ventas, key=ventas.get)  # Clave con el valor mínimo
    ventas_producto_menos_vendido = ventas[producto_menos_vendido]

    return (producto_mas_vendido, ventas_producto_mas_vendido), (producto_menos_vendido, ventas_producto_menos_vendido)

# Calcular los productos más y menos vendidos de cada tienda
producto_mas_vendido_tienda_1, producto_menos_vendido_tienda_1 = productos_mas_y_menos_vendidos(tienda_1)
producto_mas_vendido_tienda_2, producto_menos_vendido_tienda_2 = productos_mas_y_menos_vendidos(tienda_2)
producto_mas_vendido_tienda_3, producto_menos_vendido_tienda_3 = productos_mas_y_menos_vendidos(tienda_3)
producto_mas_vendido_tienda_4, producto_menos_vendido_tienda_4 = productos_mas_y_menos_vendidos(tienda_4)
```

```python
print(f"Tienda 1 - Producto más vendido: {producto_mas_vendido_tienda_1[0]} (Ventas: {producto_mas_vendido_tienda_1[1]})")
print(f"Tienda 1 - Producto menos vendido: {producto_menos_vendido_tienda_1[0]} (Ventas: {producto_menos_vendido_tienda_1[1]})")

print(f"Tienda 2 - Producto más vendido: {producto_mas_vendido_tienda_2[0]} (Ventas: {producto_mas_vendido_tienda_2[1]})")
print(f"Tienda 2 - Producto menos vendido: {producto_menos_vendido_tienda_2[0]} (Ventas: {producto_menos_vendido_tienda_2[1]})")

print(f"Tienda 3 - Producto más vendido: {producto_mas_vendido_tienda_3[0]} (Ventas: {producto_mas_vendido_tienda_3[1]})")
print(f"Tienda 3 - Producto menos vendido: {producto_menos_vendido_tienda_3[0]} (Ventas: {producto_menos_vendido_tienda_3[1]})")

print(f"Tienda 4 - Producto más vendido: {producto_mas_vendido_tienda_4[0]} (Ventas: {producto_mas_vendido_tienda_4[1]})")
print(f"Tienda 4 - Producto menos vendido: {producto_menos_vendido_tienda_4[0]} (Ventas: {producto_menos_vendido_tienda_4[1]})")
```

# 5. Envío promedio por tienda

```python
# Función para calcular el costo de envío promedio
def costo_envio_promedio(tienda):
    total_envio = 0
    cantidad_productos = 0

    # Recorrer todas las filas de la tienda
    for fila in tienda:
        total_envio += float(fila['Costo de envío'])  # Sumar el costo de envío
        cantidad_productos += 1  # Contar los productos vendidos

    # Calcular el costo promedio de envío
    if cantidad_productos > 0:  # Evitar dividir por 0
        promedio_envio = total_envio / cantidad_productos
    else:
        promedio_envio = 0

    return promedio_envio

# Calcular el costo de envío promedio para cada tienda
costo_promedio_tienda_1 = costo_envio_promedio(tienda_1)
costo_promedio_tienda_2 = costo_envio_promedio(tienda_2)
costo_promedio_tienda_3 = costo_envio_promedio(tienda_3)
costo_promedio_tienda_4 = costo_envio_promedio(tienda_4)
```

```python
print(f"Tienda 1 - Costo de envío promedio: {costo_promedio_tienda_1:.2f}")
print(f"Tienda 2 - Costo de envío promedio: {costo_promedio_tienda_2:.2f}")
print(f"Tienda 3 - Costo de envío promedio: {costo_promedio_tienda_3:.2f}")
print(f"Tienda 4 - Costo de envío promedio: {costo_promedio_tienda_4:.2f}")
```

```python
import matplotlib.pyplot as plt

# Datos
datos = [
    ['Zapp', 3.8, 32155.71, -23.5489, -46.6388],
    ['Yes Cosmetics', 4.3, 29298.19, -23.5087, -46.6205],
    ['World Tennis', 4.3, 36961.23, -23.5255, -46.5966],
    ['YouCom', 4.2, 39761.58, -23.5408, -46.6346],
    ['Yoggi', 4.2, 24114.12, -23.5631, -46.6544],
    ['Witchery', 4.4, 43217.31, -23.5243, -46.6190],
    ['Walmart', 3.5, 33893.0, -23.4780, -46.6288],
    ['Vivara', 4.3, 28219.7, -23.5581, -46.6626],
    ['Vila Romana', 4.5, 39265.61, -23.5515, -46.6556],
    ['Via Uno', 4.1, 34428.3, -23.4993, -46.6220]
]

nombres = [x[0] for x in datos]
lat = [x[3] for x in datos]
lon = [x[4] for x in datos]

plt.figure(figsize=(10, 6))
plt.scatter(lon, lat, color='orange', s=120, edgecolors='black')
for i in range(len(nombres)):
    plt.text(lon[i]+0.0008, lat[i]+0.0008, nombres[i], fontsize=8)

plt.title('Ubicación geográfica de las tiendas')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.grid(True)
plt.tight_layout()
plt.show()

plt.savefig('ubicacion_geografica_mapeado.png')
from google.colab import files
files.download('ubicacion_geografica_mapeado.png')
```
